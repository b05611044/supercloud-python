# Boston House Prices LinearRegression

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
x_data = boston.data
```

```python
y_data = boston.target
```

```python
x_data[0]
```

```python
y_data[0]
```

```python
boston.feature_names
```

```python
# 分離資料
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.3, random_state = 100)
```

```python
# 開線性回歸
regr = LinearRegression()
```

```python
# 訓練
regr.fit(x_train, y_train)
```

```python
regr.intercept_
```

```python
regr.coef_
```

# statsmodel

```python
import statsmodels.api as sm
```

```python
model = sm.OLS(y_train, x_train)
```

```python
result = model.fit()
```

```python
print(result.summary())
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
```

```python
x[0]
```

```python
y = iris.target
```

```python
y[0]
```

```python
y
```

```python
x_data = x[:, 1:3]
```

```python
x_data[0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y, test_size = 0.2, random_state = 87)
```

```python
plt.scatter(x_train[:, 0], x_train[:, 1], c = y_train)
```

```python
clf = SVC()
```

```python
clf.fit(x_train, y_train)
```

```python
y_predict = clf.predict(x_test)
```

```python
plt.scatter(x_test[:, 0], x_test[:, 1], c = y_predict)
```

```python
plt.scatter(x_test[:, 0], x_test[:, 1], c = y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_predict)
```
